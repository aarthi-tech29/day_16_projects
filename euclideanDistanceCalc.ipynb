### Project 2: Euclidean Distance Calculator 
#### Objective: Write a function to calculate the Euclidean distance between two vectors.
#### Technique: The formula is ∑(v1−v2)2. Use NumPy for efficient vectorized subtraction and squaring, and np.sqrt() for the final result 

#### Step 1: Import numpy 

```python
import numpy as np
```

##### Step 2: Define two vectors

```python
v1 = np.array([2, 3, 4]) # x,y,z - length, width, height
v2 = np.array([5, 7, 9])
```

##### Step 3: Calculate Euclidean distance

```python
distance = np.sqrt(np.sum((v1 - v2) ** 2))
distance = np.round(distance, 3)

# Euclidean Distance - Measures how far two points are apart in space
```

##### Step 4: Print result

```python
print("Vector 1:", v1)
print("Vector 2:", v2)
print("Euclidean Distance:", distance)
```
